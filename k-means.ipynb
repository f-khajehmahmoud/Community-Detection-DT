```python
from sklearn.datasets import make_blobs
import pandas as pd
import networkx as nx

dataset, classes = make_blobs(n_samples=78, n_features=2, centers=3, cluster_std=0.5, random_state=0)
# make as panda dataframe for easy understanding
data = pd.read_csv('karate.csv')
```

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,12)).fit(data)
visualizer.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0).fit(data)
```

```python
kmeans.labels_
```

```python
kmeans.inertia_
```

```python
kmeans.n_iter_
```

```python
kmeans.cluster_centers_
```

```python
from collections import Counter
Counter(kmeans.labels_)
Counter({0: 26, 1: 26, 2: 26})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(data=data, x="34", y="34.1", hue=kmeans.labels_)
plt.show()
```

```python
sns.scatterplot(data=data, x="34", y="34.1", hue=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], 
            marker="X", c="r", s=80, label="centroids")
plt.legend()
plt.show()
```
